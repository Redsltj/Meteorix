```python
from PIL import ImageEnhance
from PIL import Image



im= Image.open("Autumn_Leaves.jpg")


im_3=ImageEnhance.Contrast(im).enhance(10) #Permet de régler le contrast avec "enchance(10)" qui multiplie le contraste par 10
im_3=ImageEnhance.Color(im).enhance(10) #Permet de régler la saturation avec "enchance(10)" qui multiplie la saturation par 10
im_3=ImageEnhance.Brightness(im).enhance(10) #Permet de régler la luminosité avec "enchance(10)" qui multiplie la luminosité par 10

im_3.show()
```

```python
import numpy as np
from PIL import ImageEnhance
from PIL import Image
from astropy.io import fits
import matplotlib.pyplot as plt
import imageio


image=Image.open("20230221T104006_000080.png")
Raw=np.array(image)

Dark= np.loadtxt("correction_dark_E1000.txt")



Flat= np.loadtxt("correction_flat_E1000.txt")


Calibrated= 255*(Raw-Dark)/Flat

img = Image.fromarray(Calibrated)
bla=img.save("Orion.tiff")

im1=Image.open("Orion.tiff")

im1.show()
```
